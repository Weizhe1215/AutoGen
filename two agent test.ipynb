```python
import os
from autogen import ConversableAgent

import LLM_configs

# 定义处理旧报告的代理
agent_old_report = ConversableAgent(
    "agent_old_report",
    system_message="你负责处理旧报告的内容。",
    llm_config= LLM_configs.Qwen_strong_llm_config,
    human_input_mode="ALWAYS",
)

# 定义处理新报告的代理
agent_new_report = ConversableAgent(
    "agent_new_report",
    system_message="你负责处理新报告的内容。",
    llm_config= LLM_configs.Qwen_strong_llm_config,
    human_input_mode="ALWAYS",
)
```

```python
# 定义更新报告的生成逻辑
def update_report(old_report_relevant_part, new_report):
    initial_message = {
        "content": f"旧报告: {old_report_relevant_part} 新报告: {new_report}",
        "role": "system"
    }
    
    update_message = {
        "content": "你是一个报告审查和撰写专员，现在你有一篇旧报告和一篇新报告，你要做的是在旧报告的基础上，判断新报告中的增量信息，并将新报告中的增量信息与旧报告中的现有信息相结合，完成一篇更新的报告，如果两个报告中对某一个地方有冲突，则以新报告为准。",
        "role": "system"
    }
    
    user_message = {
        "content": "按需求生成报告，不要分段，写成一整段，要求逻辑清晰，细节完整.长度越长越好，但不要有过多的你自由发挥的解释，客观记录为主, 不要出现评价语言，例如“这体现了XXXX",
        "role": "user"
    }

    # 初始对话框内容
    initiate_message = "开始生成更新的报告"

    # 启动对话
    result = agent_old_report.initiate_chat(
        agent_new_report,
        message=initiate_message,
        messages=[initial_message, update_message, user_message]
    )

    return result

# 运行并收集反馈
old_report_relevant_part = input("请把上一篇报告的配置逻辑部分复制到此处: ")
new_report = input("请把新报告内容复制到此处: ")

# 首次生成报告
final_report = update_report(old_report_relevant_part, new_report)
print("生成的中间报告: ", final_report)

# 用户反馈和调整
feedback = input("请提供您的反馈，如果满意请输入'y'，如果不满意请输入您的调整建议: ")

while feedback.lower() != 'y':
    final_report = update_report(old_report_relevant_part, feedback)
    print("生成的中间报告: ", final_report)
    feedback = input("请提供您的反馈，如果满意请输入'y'，如果不满意请输入您的调整建议: ")

print("生成的最终报告: ", final_report)
```
